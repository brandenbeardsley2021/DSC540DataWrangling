###  Activity 1: Handling Lists
In this activity, we will generate a list of random numbers and then generate another list from the first one, which only contains numbers that are divisible by three. Repeat the experiment three times. Then, we will calculate the average difference of length between the two lists.

These are the steps for completing this activity:

Create a list of 100 random numbers.

Create a new list from this random list, with numbers that are divisible by 3.

Calculate the length of these two lists and store the difference in a new variable.

Using a loop, perform steps 2 and 3 and find the difference variable three times.

Find the arithmetic mean of these three difference values

```python
import random

# Initialize variables
differences = []
num_iterations = 3

# Perform the following steps three times
for _ in range(num_iterations):
    # Step 1: Create a list of 100 random numbers
    random_list = [random.randint(1, 1000) for _ in range(100)]

    # Step 2: Create a new list with numbers divisible by 3
    divisible_by_3 = [num for num in random_list if num % 3 == 0]

    # Step 3: Calculate the length of the two lists and find the difference
    length_random_list = len(random_list)
    length_divisible_by_3 = len(divisible_by_3)
    difference = length_random_list - length_divisible_by_3

    # Append the difference to the differences list
    differences.append(difference)

# Step 4: Calculate the arithmetic mean of the three difference values
mean_difference = sum(differences) / num_iterations

# Print the results
print("Differences:", differences)
print("Arithmetic Mean of Differences:", mean_difference)
```

In this code:

We use a loop to repeat the process three times (as specified by num_iterations).

In each iteration, we generate a list of 100 random numbers between 1 and 1000.

We create a new list called divisible_by_3 that contains only the numbers from random_list that are divisible by 3.

We calculate the lengths of both random_list and divisible_by_3 and find the difference, storing it in the difference variable.

We append each difference value to the differences list.

Finally, we calculate the arithmetic mean of the three difference values and store it in the mean_difference variable, which represents the average difference between the lengths of the two lists over the three iterations.

```python
### Step 1: Create a multiline_text variable 

multiline_text = """ It is a truth universally acknowledged, that a single man in possession of a good fortune,
must be in want of a wife. """ 
```

```python
# Step 4: Find all words in cleaned_text 
word_list = cleaned_text.split() 
```

```python
# Step 5: Create a list of unique words 
unique_words = list(set(word_list)) 
```

```python
# Step 6: Count the number of times each unique word appears 
word_count = {} 
for word in unique_words: 
    word_count[word] = word_list.count(word) 
```

```python
# Step 7: Find the top 25 words 

sorted_word_count = sorted(word_count.items(), key=lambda x:
x[1], reverse=True) 
top_25_words = sorted_word_count[:25] 
```
