```python
# Step 1: Load the necessary libraries.
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Step 2: Read in the Boston housing dataset from the local directory.
file_path = "/Users/bbeardsley/Desktop/boston_housing.csv"
df = pd.read_csv(file_path)
```

```python
# Step 3: Check the first 10 records and find the total number of records.
print("First 10 records:")
print(df.head(10))
total_records = len(df)
print(f"Total number of records: {total_records}")
```

```python
# Step 4: Create a smaller DataFrame without CHAS, NOX, B, and LSTAT.
columns_to_exclude = ['CHAS', 'NOX', 'B', 'LSTAT']
small_df = df.drop(columns=columns_to_exclude)
```

```python
# Step 5: Check the last seven records of the new DataFrame.
print("Last 7 records of the new DataFrame:")
print(small_df.tail(7))
```

```python
# Step 6: Plot histograms of all variables in the new DataFrame.
for column in small_df.columns:
    plt.hist(small_df[column], bins=20)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Step 7: Create a scatter plot of crime rate (CRIM) versus price (PRICE).
plt.scatter(small_df['CRIM'], small_df['PRICE'])
plt.title('Crime Rate vs Price')
plt.xlabel('Crime Rate (CRIM)')
plt.ylabel('Price (PRICE)')
plt.show()
```

```python
# Step 8: Create a scatter plot of log10(crime) versus price (PRICE).
plt.scatter(np.log10(small_df['CRIM']), small_df['PRICE'])
plt.title('Log10(Crime Rate) vs Price')
plt.xlabel('Log10(Crime Rate)')
plt.ylabel('Price (PRICE)')
plt.show()
```

```python
# Step 9: Calculate and print some useful statistics.
mean_rooms_per_dwelling = small_df['RM'].mean()
median_age = small_df['AGE'].median()
mean_distance_to_centers = small_df['DIS'].mean()
low_price_percentage = (len(small_df[small_df['PRICE'] < 20000]) / total_records) * 100

print(f"Mean Rooms per Dwelling: {mean_rooms_per_dwelling:.2f}")
print(f"Median Age: {median_age:.2f}")
print(f"Mean Distance to Boston Employment Centers: {mean_distance_to_centers:.2f}")
print(f"Percentage of Houses with Price < $20,000: {low_price_percentage:.2f}%")
```

Before doing any further operations, we need to use the apply method we learned in this chapter. It turns out that when reading the dataset from the CSV file, all the strings came with a whitespace character in front. So, we need to remove that whitespace from all the strings.

```python
# Create a function to strip the whitespace characters:

def strip_whitespace(s):

    return s.strip()
```

Use the apply method to apply this function to all the columns with string values, create a new column, copy the values from this new column to the old column, and drop the new column. This is the preferred method so that you don't accidentally delete valuable data. Most of the time, you will need to create a new column with a desired operation and then copy it back to the old column if necessary. Ignore any warning messages that are printed:

```python
# Education column

df_subset['education_stripped']=df['education'].apply(strip_whitespace)

df_subset['education']=df_subset['education_stripped']

df_subset.drop(labels=['education_stripped'],axis=1,inplace=True)

# Occupation column

df_subset['occupation_stripped']=df['occupation'].apply(strip_whitespace)

df_subset['occupation']=df_subset['occupation_stripped']

df_subset.drop(labels=['occupation_stripped'],axis=1,inplace=True)
```

```python
# Find the number of people who are aged between 30 and 50.

# Conditional clauses and join them by & (AND)

df_filtered=df_subset[(df_subset['age']>=30) & (df_subset['age']<=50)]

# Check the contents of the new dataset:

df_filtered.head()
```

3. Create a series and practice basic arithmetic steps

a. Series 1 = 7.3, -2.5, 3.4, 1.5

i. Index = a, c, d, e

b. Series 2 = -2.1, 3.6, -1.5, 4, 3.1

i. Index = a, c, e, f, g

c. Add Series 1 and Series 2 together and print the results

d. Subtract Series 1 from Series 2 and print the results

```python
import pandas as pd

# Creating Series 1
data_series1 = [7.3, -2.5, 3.4, 1.5]
index_series1 = ['a', 'c', 'd', 'e']
series1 = pd.Series(data_series1, index=index_series1)

# Creating Series 2
data_series2 = [-2.1, 3.6, -1.5, 4, 3.1]
index_series2 = ['a', 'c', 'e', 'f', 'g']
series2 = pd.Series(data_series2, index=index_series2)

# Adding Series 1 and Series 2
result_addition = series1 + series2
print("Result of Series 1 + Series 2:")
print(result_addition)

# Subtracting Series 1 from Series 2
result_subtraction = series2 - series1
print("\nResult of Series 2 - Series 1:")
print(result_subtraction)
```
